```python
import pandas as pd
import folium
import json
```

```python
park = pd.read_csv('./static/data/park_result2.csv', sep=',')
park.head()
```

```python
del park['Unnamed: 0']
```

```python
park.reset_index(drop=True, inplace=True)
park.head()
```

```python
len(park)
```

```python
park_dict = {}
gu = park['구별']; area = park['공원면적']
lat = park['위도']; lng = park['경도']
for i in park.index:
    park_dict[park['공원명'][i]] = [gu[i], area[i], lat[i], lng[i]]
print(park_dict)
```

```python
len(park_dict)
```

```python
for key, value in park_dict.items():
    print(value[0], value[1], value[2], value[3])
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5)
for i in park.index:
    folium.CircleMarker(
        location=[park['위도'][i], park['경도'][i]], 
        tooltip=f"{park['구별'][i]} - {park['공원명'][i]}", 
        radius= park['공원면적'][i]*0.000005,
        color='magenta', 
        fill_color='magenta'
    ).add_to(map)


title_html = '<h3 align="center" style="font-size:20px"><b>서울시 모든 공원 크기별로 표시</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))                      
map
```

```python
park_gu = pd.read_csv('./static/data/park_result2.csv')
del park_gu['Unnamed: 0']
park_gu.head()
```

```python
sum_area = park_gu.groupby(['구별'])[['공원면적']].sum()
sum_area.rename(columns={'공원면적': '총 면적'}, inplace=True)
sum_area.reset_index(inplace=True)
sum_area.head()
```

```python
count_park = park_gu.groupby(['구별'])[['공원명']].count()
count_park.rename(columns={'공원명': '공원수'}, inplace=True)
count_park.reset_index(inplace=True)
count_park.head(10)
```

```python
tmp = pd.merge(count_park, sum_area, on='구별')
tmp.tail()
```

```python
park_loc = park_gu.groupby(['구별'])[['위도', '경도']].mean()
park_loc.reset_index(inplace=True)
park_loc.head()
```

```python
tmp2 = pd.merge(tmp, park_loc, on='구별')
tmp2.head()
```

```python
area_seoul = pd.read_csv('./static/data/서울시 자치구 행정구역, 면적 통계(2019).tsv', sep='\t', thousands=',', encoding='utf-8', header=1)
area_seoul.drop([0], inplace=True)
area_seoul['면적'] = area_seoul['면적'] * 1000000
area_seoul = area_seoul[['자치구', '면적']]
area_seoul.rename({'자치구': '구별', '면적': '구면적'}, inplace=True, axis=1)
area_seoul.head()
```

```python
tmp3 = pd.merge(area_seoul, tmp2, on='구별')
tmp3.head()
```

```python
park_gu = tmp3[['구별', '구면적', '총 면적', '공원수', '위도', '경도']]
park_gu.head()
```

```python
park_gu.rename({'총 면적': '공원총면적'}, inplace=True, axis=1)
park_gu.head()
```

```python
park_gu.to_csv('./static/data/park_gu.csv', sep=',', encoding='utf8')
```
